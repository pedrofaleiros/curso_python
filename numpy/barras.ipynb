```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [1, 4, 9, 16, 25, 36]

plt.bar(x, y)
```

```python
lx = range(1, 10)
ly = []
ly_2 = []
for n in lx:
    ly.append(n*2)
    ly_2.append(n*3)

plt.bar(lx, ly, color='#cc5555')

plt.plot(lx, ly_2)
```

```python
x = [1, 3, 5]
y = [2, 6, 10]
x2 = [2, 4, 6]
y2 = [4, 8, 12]

plt.bar(x, y, color='red')
plt.bar(x2, y2, color='green')
```
